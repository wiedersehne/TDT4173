```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style('darkgrid')
```

## First Dataset

```python
seed = np.random.randint(0, 1<<32)
seed = 151143166
print(seed)
np.random.seed(seed)

C = np.array([[0.25, 0.5], [0.75, 0.5]])
y = np.random.choice(2, size=(500))
X = C[y] + 0.1 * np.random.normal(size=(len(y), 2))
X = np.where(X < 1, X, 1 - np.random.uniform(0.005, 0.1, size=(len(X), 2)))
X = np.where(X > 0, X, np.random.uniform(0.005, 0.1, size=(len(X), 2)))

data_1 = pd.DataFrame(X, columns=['x0', 'x1'])
# data_1.to_csv('data_1.csv', index=False)

plt.figure(figsize=(5, 5))
sns.scatterplot(x='x0', y='x1', data=data_1)
plt.scatter(C[:, 0], C[:, 1], color='red')
plt.xlim(0, 1), plt.ylim(0, 1);
```

```python
seed = np.random.randint(0, 1<<32)
seed = 1652758634  # 10 clusters 
seed = 2649279568  # 12 clusters
print(seed)
np.random.seed(seed)
m = 12

means = np.random.uniform(0, 1, size=(m, 2))
stds = np.random.uniform(0.05, 0.07, size=(m, 1))

selections = np.random.choice(m, size=m*100)
noise = np.random.normal(size=(len(selections), 2))
X = means[selections] + noise * stds[selections]

scale = [100, 1]
X *= scale
means *= scale

data_2 = pd.DataFrame(X, columns=['x0', 'x1'])
data_2.to_csv('data_2.csv', index=False)

plt.figure(figsize=(5, 5))
sns.scatterplot(x='x0', y='x1', data=data_2)
plt.scatter(means[:,0], means[:,1], color='red')
```

## Bonus Dataset

```python
seed = np.random.randint(0, 1<<32)
seed = 3747445327
print(seed)
np.random.seed(seed)

r = np.array([0, 0, 0, 0.65, 1])
o0 = np.array([       0,      -0.5,      +0.5, 0, 0]) * 0.4
o1 = np.array([3**0.5/4, -3**0.5/4, -3**0.5/4, 0, 0]) * 0.4
y = np.random.choice(len(r), size=1000, p=[0.05, 0.05, 0.05, 0.35, 0.5])
a = np.random.uniform(size=len(y))
e0, e1 = np.random.normal(size=(2, len(y)))

data_bonus = pd.DataFrame()
data_bonus['x0'] = o0[y] + r[y] * np.cos(2 * np.pi * a) + 0.05 * e0
data_bonus['x1'] = o1[y] + r[y] * np.sin(2 * np.pi * a) + 0.05 * e1

# data_bonus.to_csv('data_bonus.csv', index=False)

plt.figure(figsize=(5, 5))
sns.scatterplot(x='x0', y='x1', data=data_bonus)
```
